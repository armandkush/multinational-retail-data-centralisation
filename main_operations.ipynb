```python
# Extracting and clean date time data
import pandas as pd
import requests
from database_utils import DatabaseConnector
from data_extraction import DataExtractor
from data_cleaning import DataCleaning


# Create instances of the classes
data_extractor_instance = DataExtractor()
data_cleaning_instance = DataCleaning()
db_connector_instance = DatabaseConnector('db_creds.yaml')

# Link of json file
link = "https://data-handling-public.s3.eu-west-1.amazonaws.com/date_details.json"

# Calling retrieve_pdf_data method 
response = requests.get(link)
json_data = response.json()

# Load data into pandas dataframe
dt_data = pd.DataFrame(json_data)
display(dt_data)

# Clean dt_df
cleaned_dt_data = data_cleaning_instance.clean_date_time_data(dt_data)

# Uploading dim_card_details to database
upload = db_connector_instance.upload_to_db(cleaned_dt_data, 'dim_date_times')
```
